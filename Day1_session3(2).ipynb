In Python, the reason we use the from keyword when importing a specific function, class, or variable (like train_test_split) from a module or package is to indicate which part of the library we want to use. Here's why this approach is necessary.
Libraries like scikit-learn are organized into multiple modules for better organization and usability. In this case:

sklearn is the main package.
model_selection is a submodule within sklearn.
train_test_split is a function inside the model_selection submodule.
To access the function, you must specify the correct hierarchy using either:

The full path: sklearn.model_selection.train_test_split
A targeted import with from.

```python
import pandas as pd
```

This imports the pandas library and gives it an alias (pd) for convenience.
You can now use pd to access pandas functions and classes (e.g., pd.DataFrame).

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
```

This creates a dictionary named data. Each key represents a column in the dataset:
'Name' maps to a list of names: ['Alice', 'Bob', 'Charlie'].
'Age' maps to a list of ages: [25, 30, 35].
'City' maps to a list of cities: ['New York', 'Los Angeles', 'Chicago'].

```python
df = pd.DataFrame(data)
```

Converts the dictionary data into a pandas DataFrame.
A DataFrame is a 2D table-like structure with rows and columns, similar to a spreadsheet or SQL table.

```python
print("Original DataFrame:")
print(df)
```

```python
df['Salary'] = [70000, 80000, 90000]
```

Adds a new column named 'Salary' to the DataFrame df.
The values in this column are [70000, 80000, 90000].

```python
print(df)
```

Filter the DataFrame df to include only rows where the 'Age' column value is greater than 28.


```python
filtered_df = df[df['Age'] > 28]
```

Applying this mask to df results in a new DataFrame (filtered_df) containing only the rows for Bob and Charlie:



```python
print(filtered_df)
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie','David'],
    'Age': [25, 30, 35,89],
    'City': ['New York', 'Los Angeles', 'Chicago','Boston'],
    'Salary': [70000, 80000, 90000,95000]
}
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
# Check for missing values (NaN) in the DataFrame
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Rename the 'Age' column to 'Age_Years'
df = df.rename(columns={'Age': 'Age_Years'})
print(df)
```

```python
# Group data by the 'City' column and calculate the mean salary for each city
grouped_data = df.groupby('City')['Salary'].mean()
print(grouped_data)
```

```python
# Reset the index of the DataFrame (after rows have been added or removed)
df_reset = df.reset_index(drop=True)
print(df_reset)
```

```python
# Filter rows where 'Age' > 28 and 'Salary' > 75000
filtered_data = df[(df['Age_Years'] > 28) & (df['Salary'] > 75000)]
print(filtered_data)
```

```python
# Get summary statistics for numeric columns
summary_stats = df.describe()
print(summary_stats)
```

```python
# Apply a function to calculate the length of each name in the 'Name' column
df['Name_Length'] = df['Name'].apply(len)
print(df)
```

```python
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/house_prices.csv'
house_prices_data = pd.read_csv(file_path)

# 1. Display the First 5 Rows of the Dataset
first_5_rows = house_prices_data.head()
print("First 5 Rows:")
print(first_5_rows)

# 2. Display the Last 5 Rows of the Dataset
last_5_rows = house_prices_data.tail()
print("\nLast 5 Rows:")
print(last_5_rows)

# 3. Display the first N rows (e.g., first 10 rows) of the dataset
first_10_rows = house_prices_data.head(10)
print("\nFirst 10 Rows:")
print(first_10_rows)

# 4. Display the Last N Rows (e.g., last 10 rows)
last_10_rows = house_prices_data.tail(10)
print("\nLast 10 Rows:")
print(last_10_rows)

# 5. Display a specific range of rows (e.g., middle rows)
middle_rows = house_prices_data.iloc[len(house_prices_data)//2 - 5:len(house_prices_data)//2 + 5]
print("\nMiddle Rows:")
print(middle_rows)

# 6. Display Random Rows from the Dataset
random_rows = house_prices_data.sample(5)
print("\nRandom Rows:")
print(random_rows)

# 7. Display Specific Columns (e.g., 'Square_Feet', 'Price')
specific_columns = house_prices_data[['Square_Feet', 'Price']]
print("\nSpecific Columns (Square_Feet, Price):")
print(specific_columns)

# 8. Display the Data Types of the Columns
column_data_types = house_prices_data.dtypes
print("\nData Types of Columns:")
print(column_data_types)

# 9. Display Basic Statistical Summary
statistical_summary = house_prices_data.describe()
print("\nBasic Statistical Summary:")
print(statistical_summary)

# 10. Display Unique Values in the 'Square_Feet' Column
unique_values_square_feet = house_prices_data['Square_Feet'].unique()
print("\nUnique Values in 'Square_Feet' Column:")
print(unique_values_square_feet)
```
