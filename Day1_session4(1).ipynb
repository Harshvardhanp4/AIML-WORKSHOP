```python
# Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


```

Import Required Libraries
pandas: Used for handling data in a tabular format (DataFrames).
sklearn.linear_model.LinearRegression: The Linear Regression model from the scikit-learn library, which helps build and train the regression model.
matplotlib.pyplot: Used for visualizing the data and the regression line.

```python
# Define a small numeric dataset
# Example: Predict 'y' based on 'x'
data = {
    'x': [1, 2, 3, 4, 5],  # Feature
    'y': [2, 4, 5, 4, 5]   # Target
}
```

Here, we define a dictionary containing:
x (Feature): Input values or independent variable.
y (Target): The values we want to predict or dependent variable.

```python

# Convert to DataFrame
df = pd.DataFrame(data)
```

Convert the dictionary data into a pandas DataFrame for easy manipulation and analysis.

```python

# Define feature (X) and target (y)
X = df[['x']]  # Feature column
y = df['y']    # Target column
```

```python

# Create a Linear Regression model
model = LinearRegression()

```

Here, we instantiate the Linear Regression model from sklearn.
At this stage, the model is empty and needs to be trained on the dataset.

```python
# Train the model on the dataset
model.fit(X, y)

```

The .fit() method trains the Linear Regression model by finding the best-fit line for the data.
It calculates the slope (m) and intercept (c) of the line y=mx+c

```python
# Make predictions
y_pred = model.predict(X)
```

.predict(): Once trained, the model can predict the y values based on the given X values.
The predictions (y_pred) are the model's outputs for each value of X.

```python

# Display the actual and predicted values
results = pd.DataFrame({'x': df['x'], 'Actual y': y, 'Predicted y': y_pred})
print(results)

```

A new DataFrame results is created to display:
x: Input values (features).
Actual y: Original target values.
Predicted y: Model’s predicted values.

```python
# Visualize the data and the regression line
plt.scatter(X, y, color='blue', label='Actual Data')  # Scatter plot of actual data
plt.plot(X, y_pred, color='red', label='Regression Line')  # Regression line
plt.xlabel('x (Feature)')
plt.ylabel('y (Target)')
plt.title('Simple Linear Regression Example')
plt.legend()
plt.show()
```

Scatter Plot: Plots the actual data points (x vs y) in blue.
Regression Line: Plots the straight line fitted by the model in red.
Labels:
xlabel and ylabel: Descriptive labels for axes.
title: Title of the graph.
legend: Adds labels for clarity.
plt.show(): Displays the graph.

```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the larger dataset from CSV
df = pd.read_csv('/content/drive/MyDrive/house_prices_MULTIPLE.csv')

# Define the features (Square_Feet and Bedrooms) and target (Price)
X = df[['Square_Feet', 'Bedrooms']]  # Features
y = df['Price']                      # Target

# Create a Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Display the model coefficients
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

# Make predictions
y_pred = model.predict(X)

# Create a DataFrame to compare actual and predicted values
results = pd.DataFrame({
    'Square_Feet': df['Square_Feet'],
    'Bedrooms': df['Bedrooms'],
    'Actual Price': y,
    'Predicted Price': y_pred
})
print("\nComparison of Actual and Predicted Prices:")
print(results.head())

# Visualize actual vs predicted prices
plt.scatter(y, y_pred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display basic information about the dataset
print("Dataset Information:")
print(weather_data.info())

# Summary statistics of numerical columns
print("\nSummary Statistics:")
print(weather_data.describe())

# Check for missing values
print("\nMissing Values:")
print(weather_data.isnull().sum())

# Convert 'Formatted Date' to datetime format with UTC
weather_data['Formatted Date'] = pd.to_datetime(weather_data['Formatted Date'], utc=True)

# Verify the conversion
print("\nConverted 'Formatted Date':")
print(weather_data['Formatted Date'].head())

# Drop rows with missing values
weather_data_cleaned = weather_data.dropna()

# Verify the cleaning
print("\nMissing Values After Cleaning:")
print(weather_data_cleaned.isnull().sum())

# Select only numeric columns
numeric_data = weather_data_cleaned.select_dtypes(include=['float64', 'int64'])

# Check if there are numeric columns
if numeric_data.empty:
    print("\nNo numeric columns available for correlation.")
else:
    # Compute the correlation matrix
    correlation_matrix = numeric_data.corr()

    # Plot the heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap')
    plt.show()

# Filter data for rainy days
if 'Precip Type' in weather_data_cleaned.columns:
    rainy_days = weather_data_cleaned[weather_data_cleaned['Precip Type'] == 'rain']
    print("\nRainy Days Data:")
    print(rainy_days.head())
else:
    print("\n'Precip Type' column not found in the dataset.")

# Average temperature by daily summary
if 'Daily Summary' in weather_data_cleaned.columns and 'Temperature (C)' in weather_data_cleaned.columns:
    average_temp_by_summary = weather_data_cleaned.groupby('Daily Summary')['Temperature (C)'].mean()
    print("\nAverage Temperature by Daily Summary:")
    print(average_temp_by_summary)
else:
    print("\nRequired columns for average temperature calculation not found.")

# Save the cleaned data to the current directory
cleaned_file_path = 'cleaned_weather_data.csv'
weather_data_cleaned.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned data saved to: {cleaned_file_path}")

```
